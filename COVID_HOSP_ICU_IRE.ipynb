```python
import datetime
import pandas as pd
import numpy as np
import time
import seaborn as sns
from datetime import datetime
from datetime import timedelta
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
%matplotlib inline

# import convoys
# import convoys.plotting
# import convoys.regression
# import convoys.single
# import convoys.utils

pd.set_option('display.max_columns', 273)
pd.set_option('display.max_rows', 273)

df = pd.read_csv('CovidStatisticsProfileHPSCIrelandOpenData.csv')
```

```python

set1 = df.CovidCasesConfirmed.values
set2 = df.HospitalisedCovidCases.values
set3 = df.RequiringICUCovidCases.values




x1 = []
x2 = []
x3 = []


r_0_list_1 = []
r_0_list_2 = []
r_0_list_3 = []

for i in range(len(set1)):
    try:
        r_0_1 = (set1[-1-i] + set1[-2-i])/(set1[-3-i] + set1[-4-i])
    except:
        pass
    if r_0_1 == np.inf:
        r_0_1 = np.nan
    x = i*(-1)
    x1.append(x)
    r_0_list_1.append(r_0_1)
    
for i in range(len(set2)):
    try:
        r_0_2 = (set2[-1-i] + set2[-2-i])/(set2[-3-i] + set2[-4-i])
    except:
        pass
    if r_0_2 == np.inf:
        r_0_2 = np.nan
    x = i*(-1)
    x2.append(x)
    r_0_list_2.append(r_0_2)

for i in range(len(set3)):
    try:
        r_0_3 = (set3[-1-i] + set3[-2-i])/(set3[-3-i] + set3[-4-i])
    except:
        pass
    if r_0_3 == np.inf:
        r_0_3 = np.nan
    x = i*(-1)
    x3.append(x)
    r_0_list_3.append(r_0_3)
    




fig, ax = plt.subplots()
fig.set_size_inches(24, 16, forward=True)


ax.plot(x1, r_0_list_1,   'o-', label = 'Confirmed', linewidth=6)
ax.plot(x2, r_0_list_2,   'o-', label = 'Hospitalised', linewidth=6)
ax.plot(x3, r_0_list_3,   'o-', label = 'ICU', linewidth=6)



plt.rc('font', size=28)
#plt.xlabel('DATE', fontsize=20)
plt.ylabel("R_nought (2 Day Window)", fontsize=28)
plt.xlabel("Days Ago", fontsize=28)
#_ = plt.plot([0.01, 1], [0.01, 1])

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.xlim(-30 ,1)
plt.xticks(rotation=0)
plt.ylim(0,4)
#plt.yscale('log')
#plt.xscale('log')
plt.title('Infectiousness over time')

plt.legend()

plt.grid(True)
plt.show()
```
